```python
def getProducts():
    

    conn = mysql.connector.connect(
        host = "localhost",
        user = "toygarpar",
        password = "xxxxxxx",
        database = "shopclassdb"
    )

    cursor = conn.cursor()

    

    cursor.execute("SELECT * FROM products WHERE id=1")

    res = cursor.fetchone()

    print("******FETCHONE******")
    print(f"name:{res[1]}, price:{res[2]}, description:{res[4]}")
    print("--------------------")
    
    result = cursor.fetchall()
    print("******FETCHALL******")
    print(result)
    print("--------------------")


    print("******FETCHALL - FOR LOOP******")

    for product in result:
        print(product)
        print("----------------------")

    print("******FETCHALL - F format******")
    
    for product in result:
        print(f"name:{product[0]}, price:{product[1]}, description:{product[2]}")
    print("-----------------")

    conn.close()
    print("database bağlantısı kapandı")

getProducts()
```
