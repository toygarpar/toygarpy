<code style="background:yellow;color:red">Http Methods</code>


Http Get

Bilgi talebi için kullanılır. Query string ile gönderilerek server tarafından filtrelenmiş bir şekilde bilgi getirilmesini sağlar. Server tarafında bir etkisi yoktur.

allows to send http requests using python, and returns a response object with all the response data.


Http Post

Yapılan requestin body(payload) tarafı olur. Bilgi göndermek için kullanılır. Server tarafında etkisi vardır.

Request modülünü eğer bilgisayarınızda yoksa pypi.org adresinden indirebilirsiniz.

Here are the Methods available in the Requests Module:

delete(url, args)	 -  Sends a DELETE request to the specified url
get(url, params, args)	 - Sends a GET request to the specified url
head(url, args)	 - Sends a HEAD request to the specified url
patch(url, data, args)	 - Sends a PATCH request to the specified url
post(url, data, json, args)	 - Sends a POST request to the specified url
put(url, data, args)	 - Sends a PUT request to the specified url
request(method, url, args)	 - Sends a request of the specified method to the specified url

<code style="background:yellow;color:red">HTTP POST REQUEST</code>

server tarafına kayıt amacıyla bilgi göndermek

```python
import requests

#post reqest gönderildiğinde payload body içerisinde data ile gönderilir

post_talep = requests.post("https://jsonplaceholder.typicode.com/posts", data = {
    "userId": 1,
    "title": "yeni gönderi deneme deneme deneme 1-2-3",
    "body": "deneme , deneme , deneme 1-2-3 açıklaması"



} )


post_talep   #201 durum kodu - Created success status response code indicates that the request has succeeded and has led to the creation of a resource
```

```python
post_talep.text    # server tarafından assign yapılan id'e bakalım
```
