```python
#define custom serialization logic for your objects using  classes

import json

class Person:
    
    def __init__(self, name, age, email):
        self.name = name
        self.age = age
        self.email= email

    def __repr__(self):
        # Representation for developers
        return f"Person(name={self.name}, age={self.age}, email={self.email})"

    def __str__(self):
        # User-friendly representation
        return f"{self.name}, {self.age} years old, email: {self.email}"
    


class PersonEncoder(json.JSONEncoder):  # JSONEncoder class allows you to customize the encoding process.
    
    def default(self, o):               #extend the JSONEncoder and change its default method
        
        if isinstance(o, Person):
            return {"name": o.name, "age": o.age, "email": o.email}   #customize the encoding process for a custom object
        
        return super().default(o)


class PersonDecoder(json.JSONDecoder):  #to handle the deserialization of a JSON string back into a Person object
    
    def __init__(self, *args, **kwargs):

        #override the __init__ method of json.JSONDecoder and use the object_hook parameter, which allows you to define custom deserialization logic.
        
        super().__init__(object_hook=self.object_hook, *args, **kwargs) #ensures that the object_hook parameter of the parent json.JSONDecoder class is set to the custom object_hook method we defined

    def object_hook(self, obj): #object_hook is a parameter of json.JSONDecoder that allows you to define custom logic for deserializing JSON objects into Python objects

        """ 
        object_hook: This is used for post-processing JSON objects after they are decoded into Python dictionaries. It checks if the dictionary has the keys name, age, and email, and if so, it converts it into a Person object.        
        
        """


        if "name" in obj and "age" in obj and "email" in obj:
            return Person(name=obj["name"], age=obj["age"], email=obj["email"])
        return obj

```

super(): Refers to the parent class of PersonDecoder, which is json.JSONDecoder.

__init__(): This is the constructor method of the json.JSONDecoder class. It accepts various parameters, including object_hook.


object_hook=self.object_hook: We pass our custom object_hook method to the parent class's __init__ method. This ensures that every time the JSON decoder parses an object, it calls our object_hook.

```python
#create a custom object

p = Person("Toygar", 50 , "toygar@par.com")

p
```

```python
# Encode the custom object using the custom encoder

json_string = json.dumps(p, cls=PersonEncoder)

#encoded string

print(json_string)
```

```python
# Test Example
if __name__ == "__main__":
    # Create a Person object
    person = Person(name="John Doe", age=30, email="john.doe@example.com")
    
    # Serialize the Person object
    json_str = json.dumps(person, cls=PersonEncoder)
    print("Serialized JSON:", json_str)
    
    # Deserialize the JSON string back to a Person object
    person_obj = json.loads(json_str, cls=PersonDecoder)
    print("Deserialized Object:", person_obj)
```

```python
import json

class Product:
    def __init__(self, Id, ProductName, Price, Rating, Category, Options):
        self.Id = Id
        self.ProductName = ProductName
        self.Price = Price
        self.Rating = Rating
        self.Category =  Category
        self.Options = Options

    def __repr__(self):
        # Representation for developers
        return f"Product(Id={self.Id},ProductName={self.ProductName}, Price={self.Price}, Rating={self.Rating}, Category={self.Category}, Options= {self.Options})"

    def __str__(self):
        # User-friendly representation
        return f"{self.ProductName}, {self.Category} ürünü fiyatı: {self.Price}"
    


class ProductEncoder(json.JSONEncoder):  # JSONEncoder class allows you to customize the encoding process.
    
    def default(self, o):               #extend the JSONEncoder and change its default method
        
        if isinstance(o, Product):
            return {"Id": o.Id, "ProductName": o.ProductName, "Price": o.Price, "Rating":o.Rating, "Category":o.Category, "Options": o.Options}   #customize the encoding process for a custom object
        
        return super().default(o)



class ProductDecoder(json.JSONDecoder):  #to handle the deserialization of a JSON string back into a Person object
    
    def __init__(self, *args, **kwargs):

        #override the __init__ method of json.JSONDecoder and use the object_hook parameter, which allows you to define custom deserialization logic.
        
        super().__init__(object_hook=self.object_hook, *args, **kwargs) #ensures that the object_hook parameter of the parent json.JSONDecoder class is set to the custom object_hook method we defined


    def object_hook(self, obj): #object_hook is a parameter of json.JSONDecoder that allows you to define custom logic for deserializing JSON objects into Python objects

        """ 
        object_hook: This is used for post-processing JSON objects after they are decoded into Python dictionaries. It checks if the dictionary has the keys ProductName, Price, Rating, Category, Options and if so, it converts it into a Product object.        
        
        """


        if "Id" in obj and "ProductName" in obj and "Price" in obj and "Rating" in obj and "Category" in obj and "Options" in obj:
            return Product(Id=obj["Id"], ProductName=obj["ProductName"], Price=obj["Price"], Rating=obj["Rating"], Category=obj["Category"], Options= obj["Options"])
        return obj

    def dict_object_hook(self,obj):
        pass

```

```python
#********************************************
#uygulama json

import json
import os


class User:
    def __init__(self, username, password, email):
        self.username = username
        self.password = password
        self.email = email

class UserRepository:
    def __init__(self):
        self.users = []
        self.isLoggedIn = False
        self.currentUser = {}
        
        #load userrs from .json file
        self.loadUser()
        
    def loadUser(self):
        if os.path.exists("users.json"):
            with open("users.json", "r", encoding = "utf-8") as file:
                users = json.load(file)
                for user in users:
                    user = json.loads(user)
                    print(user["username"])
                    newUser = User(username= user["username"], password = user["password"], email =  user["email"])
                    self.users.append(newUser)
            print(self.users)
        
        
    def register(self, user: User):
        self.users.append(user)
        self.savetoFile()
        print("Kullanıcı Oluşturuldu.")
        
    
    
    
    def login(self, username, password):
        if self.isLoggedIn:
            print("zaten login oldunuz")
            
        else: 
            
            for user in self.users:
                if user.username == username and user.password == password:
                    self.isLoggedIn =  True
                    self.currentUser  = user
                    print("login yapıldı.")
                    break 
            
    def logout(self):
        self.isLoggedIn = False
        self.currentUserrentUser = {}
        print("Çıkış Yapıldı")
                
    def identity(self):
        if self.isLoggedIn:
            print(f" username: {self.currentUser.username}")
        else:
            print("Giriş Yapılmadı")
        
    def savetoFile(self):
        with open("users.json", "w") as file:
            list = []
            
            for user in self.users:
                
                list.append(json.dumps(user.__dict__)) #user classına dict haline çeviriyor
            
            json.dump(list, file)
    
    
repository = UserRepository()    
    
while True:
    print("Menü".center(50,"*"))
    secim = input("1- Register\n2- Login\n3- Logout\n4- Identity\n5- Exit\nSeçiminiz: ")
    if secim == "5":
        break
    else: 
        if secim == "1":
            username = input("username: ")
            password = input("password: ")
            email = input("email: ")
            
            user = User(username=username, password = password, email = email)
            repository.register(user)
            
            print(repository.users)
        elif secim == "2":
            if repository.isLoggedIn:
                print("zaten login oldunuz")
            else:
                username = input("username: ")
                password = input("password: ")
                
                repository.login(username, password)
        elif secim == "3":
            repository.logout()
        elif secim == "4":
            repository.identity()
        else:
            print("Yanlış Seçim!")
```
