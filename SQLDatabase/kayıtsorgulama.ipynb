```python
#sadece kız öğrencilerin ad ve soyad

connection = mysql.connector.connect(
        host ="127.0.0.1",
        port = 3306,
        user = "toygar",
        password = "xxxxxxx",
        database = "schooldb" )

cursor = connection.cursor()

sql = "SELECT name, Surname FROM student WHERE Gender = 'K';"


cursor.execute(sql)

try: 
    
    result = cursor.fetchall()
    
    
    print(result)
    print("*****************")
    
    for student in result:
        print(f"Name : {student[0]}, Last Name : {student[1]}")
    
    cursor.close()

except mysql.connector.Error as err:
    print("hata : ", err)




finally: 
    
    connection.close()
    print("database bağlantısı kapandı")
```
